```python
import numpy as np
import scipy.integrate as sp
import matplotlib.pyplot as plt


def volume(V,t,w_o):
    
    pi = np.pi
    
    D_m = 0.15e-10
    L = 687e-6
    R = (178e-6)/2
    V_0 = pi*R*R*L
    S_0 = 2*pi*R*L
    delta_0 = 1.915e-6
    w_other_0 = 0.1927
        
    c1 = 2*D_m*pi*L/V_0
    c2 = w_other_0*V_0
    c3 = (delta_0*S_0)/2
          
    dVdt = -(c1*V*(1-(c2/V)-w_o))/(np.log(1+(c3/V)))
    
    return dVdt


def soil_moisture(t):
    
    T = t/3600
    
    return 0.125*np.sin(np.pi*T/12)+0.325


V_0 = 0.2855*np.pi*(178e-6/2)*(178e-6/2)*(687e-6)
moisture = [soil_moisture(t) for t in range(3600*48)]
volumes = []

for time in range(3600*48):
    
    w_o = moisture[time]
    t = np.linspace(time,time+1,2)
    vol = sp.odeint(volume, V_0, t, args = (w_o,))
    volumes.append(vol[1][0])
    V_0 = vol[1][0]

fig, ax1 = plt.subplots(dpi=200)

colour = "tab:red"
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Volume (m\u00b3)", color=colour)
ax1.plot(range(3600*48), volumes, color=colour)
ax1.tick_params(axis="y", labelcolor=colour)
plt.xlim(0,3600*48)
plt.ylim(0)
plt.xticks(rotation=45)

ax2 = ax1.twinx()  
colour = "tab:blue"
ax2.set_ylabel("Mass fraction of water \n in the external environment", color=colour)  
ax2.plot(range(3600*48), moisture, color=colour)
ax2.tick_params(axis="y", labelcolor=colour)
plt.xlim(0,3600*48)
plt.ylim(0)

plt.suptitle("Diurnal variation in external moisture and tardigrade volume")

fig.tight_layout()  
plt.show()
```
